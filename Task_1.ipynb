```python
from langchain.embeddings import HuggingFaceEmbeddings    ### Import Libraries
from langchain.vectorstores import Chroma
from langchain.text_splitter import CharacterTextSplitter
from langchain.docstore.document import Document
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import os
from langgraph.graph import StateGraph, END
```

```python
import langgraph
print(dir(langgraph))
print(dir(langgraph.graph))
```

```python
knowledge_text = """
Renewable energy comes from natural sources such as sunlight, wind, rain,
tides, waves, and geothermal heat. Benefits include reducing pollution,
slowing climate change, creating jobs, and reducing energy imports.
Solar power uses photovoltaic cells to convert sunlight into electricity.
Wind turbines generate power by spinning blades connected to generators.
"""

with open("data.txt", "w") as f:
    f.write(knowledge_text)
```

```python
# Load and Store Data in ChromaDB
with open("data.txt", "r") as f:
    data = f.read()

text_splitter = CharacterTextSplitter(chunk_size=300, chunk_overlap=50)
docs = text_splitter.split_text(data)
documents = [Document(page_content=t) for t in docs]

# Create embeddings
embedding_model = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Store in Chroma DB
vectordb = Chroma.from_documents(documents, embedding_model)
```

```python
llm = pipeline("text-generation", model="gpt2", max_new_tokens=150)
```

```python
def plan(state):
    query = state["query"]
    print("\n[PLAN] Received Query:", query)
    return {"query": query, "need_retrieval": True}

def retrieve(state):
    if state["need_retrieval"]:
        docs = vectordb.similarity_search(state["query"], k=3)
        print("\n[RETRIEVE] Retrieved Context:")
        for d in docs:
            print("-", d.page_content)
        return {"documents": docs, "query": state["query"]} # Include query in the returned state
    return {"documents": [], "query": state["query"]} # Include query in the returned state

def answer(state):
    print("\n[ANSWER] State received:", state) # Add this line to inspect the state
    context = "\n".join([d.page_content for d in state["documents"]])
    prompt = f"Answer using context:\n{context}\n\nQuestion: {state['query']}"
    model_output = llm(prompt)[0]['generated_text']
    print("\n[ANSWER] Generated Answer:", model_output)
    return {"answer": model_output, "query": state["query"]} # Include query in the returned state

def reflect(state):
    ans = state["answer"]
    query = state["query"]
    score = 1 if query.lower().split()[0] in ans.lower() else 0
    print("\n[REFLECT] Relevance Score:", score)
    return {"score": score, "answer": ans} # Include answer in the returned state
```

```python
graph = StateGraph(dict)

graph.add_node("plan", plan)
graph.add_node("retrieve", retrieve)
graph.add_node("answer", answer)
graph.add_node("reflect", reflect)

graph.set_entry_point("plan")
graph.add_edge("plan", "retrieve")
graph.add_edge("retrieve", "answer")
graph.add_edge("answer", "reflect")
graph.add_edge("reflect", END)

app = graph.compile()
```

```python
user_question = input("Ask a question: ")

output = app.invoke({"query": user_question})
print("\n\n FINAL OUTPUT:")
print("Answer:", output["answer"])
print("Score:", output["score"])
```

**Project report**

RAG-Based Q&A AI Agent

This project demonstrates a Retrieval-Augmented Generation (RAG) pipeline using LangGraph, LangChain, ChromaDB and HuggingFace models. The agent accepts a user question, retrieves relevant context from a local knowledge base, and generates a final answer using an LLM. The workflow is divided into four LangGraph nodes: plan (understand the query), retrieve (vector similarity search over ChromaDB), answer (LLM response using retrieved context), and reflect (validate relevance of the generated answer). The system prints logs from each step, allowing users to trace exactly how the agent thinks and how the final answer was produced. The entire solution runs fully in Google Colab and uses lightweight models (MiniLM embeddings + GPT-2) for fast CPU-only execution.

During development, the main challenges involved dependency compatibility across LangChain, LangGraph and ChromaDB, fine-tuning text chunk size for optimal retrieval, and minimizing hallucination in the LLM output. To solve these issues, character-based splitting was used, embeddings were stored locally in ChromaDB, and a simple reflection mechanism was added to score answer relevance. The final result is a working RAG-based agent that demonstrates real-world AI Agent workflow concepts such as retrieval, context-aware generation, and validation. Future improvements could include a Streamlit UI, larger LLM models for better output quality, and automated evaluation using ROUGE/BERTScore or LLM-as-a-judge.
